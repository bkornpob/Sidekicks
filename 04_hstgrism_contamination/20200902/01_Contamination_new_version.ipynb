# Story
We would like to extend the functionality of hstgrism.contamination.Contamination class by adding the computation for cps_contaminate as demonstrated in 20200901.

# Previously
20200901 continued with removing X01 contamination from GRB190829A. It developed a workflow that can extend Contamination class.


```python
from astropy.io import fits
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter
from hstgrism.container import Container
from hstgrism.contamination import Contamination
import copy
```

```python
trace_object_csv = '/Users/kbhirombhakdi/_work/GRB190829A/02_bkg/drzblot_GRB190829A_201909_G102-rerun/drzblot_GRB190829A_201909_G102-rerun_trace.csv'
trace_contaminate_csv = '../20200828/01_outputs/X01_trace.csv'
halfdy_object = pd.read_csv('/Users/kbhirombhakdi/_work/GRB190829A/02_bkg/drzblot_GRB190829A_201909_G102-rerun/drzblot_GRB190829A_201909_G102-rerun_tbox.csv').halfdyup.values
instrument = 'HST-WFC3-IR-G102'
t = pd.read_csv('../20200826/X01_together_G102/X01_together_G102.csv')
contaminate_spectrum_wavelength = t.ww.values
contaminate_spectrum_flam = t.flam_smooth.values
t = fits.open('/Users/kbhirombhakdi/_work/_calib_files/WFC3.IR.G102.cal.V4.32/WFC3.IR.G102.1st.sens.2.fits')[1].data
sensitivity_curve_wavelength = t.WAVELENGTH
sensitivity_curve_value = t.SENSITIVITY
do_drzblot_apcorr = True
t = pd.read_csv('/Users/kbhirombhakdi/_work/GRB190829A/04_drzblot_crclean_apcorr/drzblot_S01_201909_G102/drzblot_S01_201909_G102_halfdy5_apcorr.csv')
drzblot_apcorr_ww = t.ww.values.copy()
drzblot_apcorr_value = savgol_filter(t.apcorr.values.copy(),window_length=21,polyorder=1)
contaminationobj = Contamination(trace_object_csv,trace_contaminate_csv,halfdy_object,instrument,
              contaminate_spectrum_wavelength,contaminate_spectrum_flam,
              sensitivity_curve_wavelength,sensitivity_curve_value,
              do_drzblot_apcorr,drzblot_apcorr_ww,drzblot_apcorr_value,
             )
contaminationobj.compute()
contaminationobj.show()
```

```python
cps_contaminate = contaminationobj.combine_df.cps_contaminate.values.copy()
plt.plot(cps_contaminate,'x-')
plt.ylim(-1.,1.)
m = np.argwhere(cps_contaminate < 0.).flatten()
print('cps_contaminate has positive/negative spikes at index around {0}. We replace these values by zero.'.format(m.max()))
```

```python
cps_contaminate[:m.max()+1] = 0.
plt.plot(cps_contaminate)
contaminationobj_new = copy.deepcopy(contaminationobj)
contaminationobj_new.combine_df.cps_contaminate = cps_contaminate.copy()
```

```python
##### show #####
tx = contaminationobj_new.combine_df.xg_object.values
t = pd.read_csv('/Users/kbhirombhakdi/_work/GRB190829A/05_flam_with_drzblot_correction/drzblot_GRB190829A_201909_G102-rerun/drzblot_GRB190829A_201909_G102-rerun-rerun_flam.csv')
ty = t.cps.values
plt.plot(tx,ty,'b:',label='GRB190829A with contamination')
plt.plot(tx,ty - contaminationobj_new.combine_df.cps_contaminate.values,'b-',label='GRB190829A clean')
plt.plot(tx,contaminationobj_new.combine_df.cps_contaminate.values,'k-',label='X01 contaminate')
plt.xlabel('pixX')
plt.ylabel('cps halfdy = 5')
plt.legend(loc=[1.01,0.])
```
