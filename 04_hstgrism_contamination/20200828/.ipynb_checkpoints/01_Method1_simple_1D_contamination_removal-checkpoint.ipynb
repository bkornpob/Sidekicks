# Task
- Get X01 trace in GRB frame.
- Given GRB cutout:
 - Given xcut:
  - Get ycut_grb, ycut_x01, halfdy_grb
  - Define yap_grb_low = ycut_grb - halfdy_grb, yap_grb_up = ycut_grb + halfdy
  - Let's assume ycut_x01 > ycut_grb (i.e., X01 is above GRB):
   - if both yap_grb_up and yab_grb_low < ycut_x01:
     - Get apcorr_x01_in at halfdy_x01_in = ycut_x01 - yap_grb_up (becareful to EXCLUDE GRB aperture in this region)
     - Get apcorr_x01_out at halfdy_x01_out = ycut_x01 - yap_grb_low (becareful to INCLUDE GRB aperture in this region)
     - Compute contamination_fraction_x01 = (apcorr_x01_out - apcorr_x01_in) / 2.
   - Make code possible for other variations?
  - Make code possible for other variations?
- Convert X01 flam to cps_total, then convolve with contamination_fraction_x01. That is, x01_cps_contamination = x01_cps_total * contamination_fraction_x01
- Use GRB cps_total (or convert from flam), and subtract X01 contribution, i.e., grb_cps_total_x01_removed = grb_cps_total - x01_cps_contamination.
- Convert grb_cps_total_x01_removed back to flam.


```python
from hstphot.show_sources import show_sources
from hstphot.read_ds9region import read_ds9region
from hstphot.container import Container
from hstgrism.computexyref import ComputeXYREF
from hstgrism.wfc3irg102 import WFC3IRG102
from hstgrism.contamination import Contamination
from astropy.io import fits
from scipy.signal import savgol_filter
from scipy.interpolate import interp1d
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import copy
confile = '/Users/kbhirombhakdi/_work/_calib_files/WFC3.IR.G102.cal.V4.32/G102.F110W.V4.32.conf'
beam = 'A'
window_length = 11
polyorder = 0
wavebound = (7500.,11500.)
```

```python
dfile = ('/Users/kbhirombhakdi/_work/_data/15089/mastDownload/HST/idlk01jjq/idlk01jjq_flt.fits',1)
ds9regfile = '/Users/kbhirombhakdi/_work/_data/15089/mastDownload/HST/idlk01jjq/idlk01jjq.reg'
drzblotfile = ('/Users/kbhirombhakdi/_work/GRB190829A/01_making_drzblot/drzblot_GRB190829A_201909_G102/idlk01jmq_flt.fits',1)
grb_trace_file = '/Users/kbhirombhakdi/_work/GRB190829A/02_bkg/drzblot_GRB190829A_201909_G102-rerun/drzblot_GRB190829A_201909_G102-rerun_trace.csv'
grb_tbox_file = '/Users/kbhirombhakdi/_work/GRB190829A/02_bkg/drzblot_GRB190829A_201909_G102-rerun/drzblot_GRB190829A_201909_G102-rerun_tbox.csv'
x01_trace_file = None # will generate below
grb_flam_csv = '/Users/kbhirombhakdi/_work/GRB190829A/05_flam_with_drzblot_correction/drzblot_GRB190829A_201909_G102-rerun/drzblot_GRB190829A_201909_G102-rerun-rerun_flam.csv'
x01_flam_csv = '/Users/kbhirombhakdi/_work/Sidekicks/20200826/X01_together_G102/X01_together_G102.csv'

```

```python
string = '01_outputs'
containerobj = Container(string,string,'pdf',False)
sourcelist = read_ds9region(ds9regfile)
for i in sourcelist:
    sourcelist[i] = sourcelist[i] - np.array((1.,1.))
show_sources(fits.open(dfile[0])[dfile[1]].data,sourcelist,save_large=True,save_zoom=True,container=containerobj)
```

```python
t = ComputeXYREF(sourcelist['X01'],'default','default',confile,beam,drzblotfile,dfile)
t.compute()
tc = copy.deepcopy(containerobj)
tc.data['saveprefix'] = 'X01'
x01_gobj = WFC3IRG102(confile,beam,t.data['xyref'],container=tc)
x01_gobj.compute()
x01_gobj.save()
```

```python
x01_trace_file = './{0}/{1}_trace.csv'.format(tc.data['savefolder'],tc.data['saveprefix'])
trace_x01 = pd.read_csv(x01_trace_file)
trace_grb = pd.read_csv(grb_trace_file)
##### show #####
t = fits.open(drzblotfile[0])[drzblotfile[1]].data
plt.imshow(t,origin='lower',cmap='viridis',vmin=0,vmax=2)
##### X01 #####
tx = np.ceil(trace_x01.xh.values + trace_x01.xyref[0]).astype(int)
ty = np.ceil(trace_x01.yh.values + trace_x01.xyref[1]).astype(int)
m = np.argwhere((tx == tx.min()) | (tx == tx.max()))
plt.scatter(tx[m],ty[m],s=10,edgecolor='purple',facecolor='purple',marker='s',label='X01')
plt.plot(tx[m],ty[m],color='purple',ls=':',alpha=0.2,lw=4)
##### GRB #####
tx = np.ceil(trace_grb.xh.values + trace_grb.xyref[0]).astype(int)
ty = np.ceil(trace_grb.yh.values + trace_grb.xyref[1]).astype(int)
m = np.argwhere((tx == tx.min()) | (tx == tx.max()))
plt.scatter(tx[m],ty[m],s=10,edgecolor='red',facecolor='red',marker='x',label='GRB190829A')
plt.plot(tx[m],ty[m],'r:',alpha=0.2,lw=4)
##### ending #####
dx,dy = 15,15
plt.xlim(tx.min()-dx,tx.max()+dx)
plt.ylim(ty.min()-dy,ty.max()+dy)
plt.legend(fontsize=8,loc=(1.,1.))
plt.tight_layout()
```

```python
halfdy_grb = pd.read_csv(grb_tbox_file).halfdyup.values.copy()
instrument = 'HST-WFC3-IR-G102'
contamination_obj = Contamination(grb_trace_file,x01_trace_file,halfdy_grb,instrument=instrument)
contamination_obj.compute()
contamination_obj.save(containerobj)
contamination_obj.show(save=True,container=containerobj)
```

# New chunk

```python
grb_flam = pd.read_csv(grb_flam_csv)
x01_flam = pd.read_csv(x01_flam_csv)
##### show #####
plt.plot(grb_flam.ww,grb_flam.cps / grb_flam.apcorr,label='GRB190829A')
ty = x01_flam.cps / x01_flam.apcorr
plt.plot(x01_flam.ww,ty,'b:')
plt.plot(x01_flam.ww,savgol_filter(ty,window_length=window_length,polyorder=polyorder),'b-',label='X01')
plt.legend(loc=(1.01,0.))
plt.xlabel('obs. wavelength (A)',fontsize=12)
plt.ylabel('cps_total',fontsize=12)
plt.tight_layout()
string = './{0}/{1}_cps.pdf'.format(containerobj.data['savefolder'],containerobj.data['saveprefix'])
plt.savefig(string,plotformat='pdf',bbox_inches='tight')
print('Save {0}'.format(string))
```

```python
tx = x01_flam.ww.values
ty = x01_flam.cps.values / x01_flam.apcorr.values
cps_model = interp1d(tx,savgol_filter(ty,window_length=window_length,polyorder=polyorder))
tx_in_grb = contamination_obj.combine_df.ww_contaminate.values.astype(float)
ty_in_grb = cps_model(tx_in_grb)
combine_df = contamination_obj.combine_df.copy()
combine_df['cps_total_contaminate'] = ty_in_grb.copy()
combine_df['cps_total_object'] = grb_flam.cps / grb_flam.apcorr
##### show #####
plt.figure(figsize=(8,8))
tx,ty = combine_df['xg_object'],combine_df['cps_total_object']
plt.plot(tx,ty,'k:',label='GRB190829A total')
tyy = combine_df['cps_total_contaminate']
plt.plot(tx,tyy,'b:',label='X01 total')
tyy = tyy * (combine_df['apcorr_contaminate_out'] - combine_df['apcorr_contaminate_in'])
plt.plot(tx,tyy,'b-',label='X01 contamination')
tyy = tyy.astype(float)
tyy[np.isnan(tyy)] = 0.
ty_sub = ty - tyy
plt.plot(tx,ty_sub,'k-',label='GRB190829A sub')
plt.ylabel('cps_total',fontsize=12)
plt.xlabel('pixX',fontsize=12)
plt.legend(loc=(1.01,0.))
plt.tight_layout()
string = './{0}/{1}_cspsub.pdf'.format(containerobj.data['savefolder'],containerobj.data['saveprefix'])
plt.savefig(string,plotformat='pdf',bbox_inches='tight')
print('Save {0}'.format(string))
```
