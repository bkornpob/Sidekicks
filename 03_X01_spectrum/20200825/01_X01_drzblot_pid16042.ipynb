```python
from hstgrism.headersummary import HeaderSummary
from hstgrism.container import Container
from astropy.io import fits
import pandas as pd
import glob
```

```python
fltlist = glob.glob('/Users/kbhirombhakdi/_work/_data/16042/HST/*/*flt.fits')
tl = []
for ii,i in enumerate(fltlist):
    t = HeaderSummary((fltlist[ii],1))
    t.fetch()
    tl.append(t)
headerlist = HeaderSummary(None).combine(tl)
    
```

```python
t = pd.DataFrame(headerlist)
t.sort_values(by = ['EXPSTART'],inplace=True)
t
```

```python
tt = t.to_dict(orient='list')
identifier = '01_16042_hsum'
containerobj = Container(identifier,identifier,'pdf','False')
HeaderSummary(None).save(containerobj,tt)
```
