# Story
We would like to arbitrarily adjust X01 trace for a better extraction.

# Previously
20200824/01 extracted X01 spectrum using HSTGRISM output from 20200821. The trace looked slightly offset causing the need for halfdy >= 4 to avoid significant error from the trace alignment. Using halfdy >= 3 also caused X01 picking up C02 contamination.

# Resources
- HSTGRISM products: 20200821/X01_201911_G102_resurrected/*
- HSTPHOT product: 20200821/X01_mag.csv (this file was produced after the request from this task).

# Note:
- For G102, halfdy and amount of light collected:
 - 3, 90%
 - 6, 95%
- We noticed background gradient in the cutout region. The gradient showed brighter on the light. From examining the full frame, this gradient tends to be a diffused one starting from the middle of the frame, and fanning out.
 - This can still be seen the subtraction with norder = 0. We might need a better estimation.


```python
from hstgrism.container import Container
from hstgrism.objectmask import ObjectMask
from hstgrism.grismapcorr import GrismApCorr
from hstgrism.show_sum1d import show_sum1d
from hstphot.mag2flux import mag2flux
from astropy.io import fits
from scipy.interpolate import interp1d
from scipy.signal import savgol_filter as savgol
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

wavebound = {'G102':(8000.,11500.)}
yh_adjust = 1.
```

```python
##########
# Container
##########
sourcename = 'X01'
dateobs = '201911'
filterobs = 'G102'
idadjust = '{0}_{1}_{2}_adjust'.format(sourcename,dateobs,filterobs)
containerobj = Container(identifier,idadjust,'pdf',False)
```

```python
##########
# read files
##########
identifier = '{0}_{1}_{2}_resurrected'.format(sourcename,dateobs,filterobs)
gfile = ('../20200821/{0}/{0}_cutout.fits'.format(identifier),1)
bfile = ('../20200821/{0}/{0}_cutbkg.fits'.format(identifier),1)
gdata = fits.open(gfile[0])[gfile[1]].data
bdata = fits.open(bfile[0])[bfile[1]].data
trace = pd.read_csv('../20200821/{0}/{0}_trace.csv'.format(identifier))
bbcorner = pd.read_csv('../20200821/{0}/{0}_bbcorner.csv'.format(identifier))
bbox = pd.read_csv('../20200821/{0}/{0}_bbox.csv'.format(identifier))
```

```python
##########
# make different mask size
##########
halfdy_max = 10
halfdy_list = np.arange(halfdy_max)
ny,nx = gdata.shape
tx = trace.xh.values + trace.xyref.values[0] - bbcorner.bb0x.values[0]
ty = trace.yh.values + trace.xyref.values[1] - bbcorner.bb0y.values[0] + yh_adjust
objmask_list = []
for ii,i in enumerate(halfdy_list):
    t = ObjectMask(nx,ny,tx,ty,i,i)
    t.compute()
    objmask_list.append(t)
##### show #####
# for ii,i in enumerate(objmask_list):
#     plt.figure(),plt.imshow(objmask_list[ii].mask,origin='lower')
```

```python
##########
# cps
##########
cps_list = []
for ii,i in enumerate(objmask_list):
    mdata = objmask_list[ii].mask.copy()
    exdata = (gdata - bdata) * mdata
    t = np.nansum(exdata,axis=0)[bbox.padxleft[0]:-bbox.padxright[0]]
    cps_list.append(t)
##### show #####
plt.figure(figsize=(10,10))
for ii,i in enumerate(objmask_list):
    plt.plot(cps_list[ii],label='halfdy {0}'.format(halfdy_list[ii]))
plt.legend(bbox_to_anchor=(1.,1.),fontsize=10)
plt.xlabel('pixX',fontsize=12)
plt.ylabel('cps',fontsize=12)
```

```python
##### show #####
for ii,i in enumerate(objmask_list):
    plt.figure()
    m = np.isfinite(gdata)
    vmin,vmax = np.percentile(gdata[m],5.),np.percentile(gdata[m],90.)
    plt.imshow(gdata,origin='lower',cmap='viridis',vmin=vmin,vmax=vmax)
    plt.imshow(objmask_list[ii].mask,origin='lower',cmap='Greys',alpha=0.2)
    
```

# Note
From the adjustment, now we have good aperture when halfdy >= 2. With halfdy <= 4, C02 contamination looks insignificant.


```python
##########
# apcorr
##########
instrument = 'HST-WFC3-IR-G102'
apcorr_list = []
for ii,i in enumerate(objmask_list):
    apsizepix = np.full_like(trace.ww.values,halfdy_list[ii]*2 + 1,dtype=float)
    apsizearcsec = apsizepix * GrismApCorr().table[instrument]['scale']
    t = GrismApCorr(instrument=instrument,apsize=apsizearcsec,wave=trace.ww.values,aptype='diameter',apunit='arcsec',waveunit='A')
    t.compute()
    apcorr_list.append(t.data['apcorr'])
##########
# sensitivity
##########
sfile = ('/Users/kbhirombhakdi/_work/_calib_files/WFC3.IR.G102.cal.V4.32/WFC3.IR.G102.1st.sens.2.fits',1)
sdata = pd.DataFrame(fits.open(sfile[0])[sfile[1]].data)
smodel = interp1d(sdata.WAVELENGTH,sdata.SENSITIVITY,kind='linear',bounds_error=False,fill_value=None)
sensitivity = smodel(trace.ww.values)
sensitivity[~np.isfinite(sensitivity)] = 1. # replace nan with 1. to avoid divided-by-zero problem
##########
# wwperpix
##########
xhdiff = np.diff(trace.xh.values)
xhdiff = np.append(xhdiff,xhdiff[-1]) # add one last element to maintain the array dimension
wwdiff = np.diff(trace.ww.values)
wwdiff = np.append(wwdiff,wwdiff[-1]) # add one last element to maintain the array dimension
wwperpix = wwdiff/xhdiff

```

```python
##########
# drzblot apcorr
##########
drzblot_apcorr_list = []
for ii,i in enumerate(objmask_list):
    th = min(halfdy_list[ii],7)
    string = '../../GRB190829A/04_drzblot_crclean_apcorr/drzblot_S01_{0}_{1}/drzblot_S01_{0}_{1}_halfdy{2}_apcorr.csv'.format(dateobs,filterobs,th)
    t = pd.read_csv(string)
    model = interp1d(t.ww.values,t.apcorr.values,kind='nearest',bounds_error=False,fill_value='extrapolate')
    tt = model(trace.ww.values)
    drzblot_apcorr_list.append(tt)
##### show #####
plt.figure()
for ii,i in enumerate(objmask_list):
    plt.plot(trace.ww.values,drzblot_apcorr_list[ii],label='halfdy {0}'.format(halfdy_list[ii]))
plt.legend()
    
```

```python
##########
# flam
##########
flam_list = []
for ii,i in enumerate(objmask_list):
    drzblot_apcorr = drzblot_apcorr_list[ii]
    apcorr = apcorr_list[ii]
    cps = cps_list[ii]
    t = cps / (apcorr * sensitivity * wwperpix * drzblot_apcorr)
    flam_list.append(t)
##### show #####
plt.figure()
for ii,i in enumerate(objmask_list):
    tx = trace.ww.values
    ty = flam_list[ii]
    m = np.argwhere((tx >= wavebound[filterobs][0])&(tx <= wavebound[filterobs][1])).flatten()
    plt.plot(tx[m],ty[m],label='halfdy {0}'.format(halfdy_list[ii]))
plt.legend(bbox_to_anchor=(1.,1.),fontsize=10)
plt.xlabel('obs. wavelength (A)',fontsize=12)
plt.ylabel('flam',fontsize=12)


    
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax1 = plt.subplots(figsize=(10,10))
for ii,i in enumerate(objmask_list):
    x = trace.ww.values
    y = flam_list[ii]
    m = np.argwhere((x >= wavebound[filterobs][0])&(x <= wavebound[filterobs][1])).flatten()
    ax1.plot(x, y,label='halfdy {0}'.format(halfdy_list[ii]))
ax1.set_xlim(x.min(), x.max())
ax1.set_ylim(1e-18,3e-17)
ax1.set_xlabel('obs. wavelength (A)',fontsize=12)
ax1.set_ylabel('flam',fontsize=12)
ax2 = ax1.twiny()                    # make another frame that shares y-axis
ax2.set_xlim(0,len(x)) # set the y-range limit
ax2.set_title('pixX',fontsize=12)
fig.legend(fontsize=10)
fig.tight_layout()
```

# Output

```python
for ii,i in enumerate(objmask_list):
    ww = trace.ww.values
    cps = cps_list[ii]
    apcorr = apcorr_list[ii]
    drzblot_apcorr = drzblot_apcorr_list[ii]
    flam = flam_list[ii]
    output = {'ww':ww,'cps':cps,'apcorr':apcorr,'drzblot_apcorr':drzblot_apcorr,'sensitivity':sensitivity,'wwperpix':wwperpix,'flam':flam}
    string = './{1}/flam_halfdy{0}.csv'.format(halfdy_list[ii],idadjust)
    pd.DataFrame(output).to_csv(string)
    print('Save {0}'.format(string))
    
```
