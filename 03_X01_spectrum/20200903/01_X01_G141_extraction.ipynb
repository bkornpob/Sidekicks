```python
from hstphot.read_ds9region import read_ds9region
from hstgrism.container import Container
from hstgrism.computexyref import ComputeXYREF
from hstgrism.wfc3irg141 import WFC3IRG141
from hstgrism.show_overview import show_overview
from hstgrism.cutout2d import Cutout2D
from hstgrism.show_bbox import show_bbox
from hstgrism.background import Background
from hstgrism.to_fullframe import to_fullframe
from hstgrism.objectmask import ObjectMask
from hstgrism.grismapcorr import GrismApCorr
import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt
import pandas as pd
from scipy.interpolate import interp1d
from scipy.signal import savgol_filter
from hstphot.mag2flux import mag2flux
```

```python
config = {'201911':{'gfile':('/Users/kbhirombhakdi/_work/Sidekicks/03_X01_spectrum/20200824/drzblot_X01_201911_G141/idwbk4d0q_flt.fits',1),
                    'dfile':('/Users/kbhirombhakdi/_work/_data/15510/mastDownload/HST/idwbk4czq/idwbk4czq_flt.fits',1),
                    'ds9file':'/Users/kbhirombhakdi/_work/_data/15510/mastDownload/HST/idwbk4czq/idwbk4czq.reg',
                    'confile':'/Users/kbhirombhakdi/_work/_calib_files/WFC3.IR.G141.cal.V4.32/G141.F140W.V4.32.conf',
                    'container':Container('X01_201911_G141','X01_201911_G141','pdf',False),
                    'xy_adjust':np.array([1.,0.]),
                   },
          '202001':{'gfile':('/Users/kbhirombhakdi/_work/Sidekicks/03_X01_spectrum/20200824/drzblot_X01_202001_G141/idwbk6m7q_flt.fits',1),
                    'dfile':('/Users/kbhirombhakdi/_work/_data/15510/mastDownload/HST/idwbk6m6q/idwbk6m6q_flt.fits',1),
                    'ds9file':'/Users/kbhirombhakdi/_work/_data/15510/mastDownload/HST/idwbk6m6q/idwbk6m6q.reg',
                    'confile':'/Users/kbhirombhakdi/_work/_calib_files/WFC3.IR.G141.cal.V4.32/G141.F140W.V4.32.conf',
                    'container':Container('X01_202001_G141','X01_202001_G141','pdf',False),
                    'xy_adjust':np.array([0.,1.]),
                   },
         }
```

```python
grismobj = {}
for ii,i in enumerate(config):
    xyd = read_ds9region(config[i]['ds9file'])['X01'] - config[i]['xy_adjust']
    t = ComputeXYREF(xyd,'default','default',config[i]['confile'],'A',config[i]['gfile'],config[i]['dfile'],config[i]['container'])
    t.compute()
    t.save()
    xyref = t.data['xyref']
    t = WFC3IRG141(config[i]['confile'],'A',xyref,container=config[i]['container'])
    t.compute()
    t.save()
    grismobj[i] = t
    
```

```python
for ii,i in enumerate(config):
    xyd = read_ds9region(config[i]['ds9file'])['X01'] - config[i]['xy_adjust']
    show_overview(config[i]['gfile'],config[i]['dfile'],xyd,grismobj[i].output['xyref'],grismobj[i].output['xh'],grismobj[i].output['yh'],grismobj[i].output['ww'],objname='X01_{0}_G141'.format(i),save=True,container=config[i]['container'])
    
```

```python
cutoutobj = {}
for ii,i in enumerate(config):
    gfile = config[i]['gfile']
    trace = grismobj[i].output
    xg = trace['xh'] + trace['xyref'][0]
    yg = trace['yh'] + trace['xyref'][1]
    halfdy = 10
    tbox = {'xg':xg,'yg':yg,'halfdyup':halfdy,'halfdylow':halfdy}
    bbox={'padxleft': 5, 'padxright': 5, 'padyup': 10, 'padylow': 5}
    image = fits.open(gfile[0])[gfile[1]].data
    do_mask = True
    t = Cutout2D(tbox,bbox,image,config[i]['container'])
    t.compute(do_mask)
    t.save()
    ##### show #####
#     maskdata = t.mask
    maskdata = None
    plt.figure()
    show_bbox(t.cutout,objname='X01_{0}_G141'.format(i),maskdata=maskdata,save=True,container=config[i]['container'])
    ##### keep #####
    cutoutobj[i] = t
    
```

```python
bkgobj = {}
for ii,i in enumerate(config):
    t = cutoutobj[i]
    tt = Background(gdata=t.cutout,norder=1,mdata=t.mask,container=config[i]['container'])
    m = tt.poly2d.data['MASK'].copy()
    m[0:15,0:60] = True
    m[27:,0:90] = True
    tt.poly2d.model['MASKFIT'] = m.copy()
    bkgobj[i] = tt
    plt.figure()
    td = tt.poly2d.data['Y']
    m = np.isfinite(td)
    vmin,vmax = np.percentile(td[m],5.),np.percentile(td[m],90.)
    plt.imshow(td,origin='lower',cmap='viridis',vmin=vmin,vmax=vmax)
    plt.imshow(tt.poly2d.model['MASKFIT'],origin='lower',cmap='Greys',alpha=0.2,vmin=0)
```

```python
for ii,i in enumerate(config):
    t = bkgobj[i]
    t.poly2d.model['NORDER']=0
    t.poly2d.fit()
    t.save()
    bkgobj[i] = t
    to_fullframe(cutdata=t.poly2d.model['YFIT'],fullfile=config[i]['gfile'],savesuffix='bkgfull',container=config[i]['container'],**cutoutobj[i].bbcorner)
    to_fullframe(cutdata=t.poly2d.data['Y'] - t.poly2d.model['YFIT'],fullfile=config[i]['gfile'],savesuffix='subfull',container=config[i]['container'],**cutoutobj[i].bbcorner)
    ##########
    # show
    ##########
    tt = t.poly2d.data['Y']
    m = np.isfinite(tt)
    vmin,vmax = np.percentile(tt[m],5.),np.percentile(tt[m],95.)
    fig = plt.figure(figsize=(15,3))
    ax1 = fig.add_subplot(1,3,1)
    ax2 = fig.add_subplot(1,3,2)
    ax3 = fig.add_subplot(1,3,3)
    y = t.poly2d.data['Y']
    yfit = t.poly2d.model['YFIT']
    ysub = y - yfit
    ax1.imshow(y,origin='lower',cmap='viridis',vmin=vmin,vmax=vmax)
    ax2.imshow(yfit,origin='lower',cmap='viridis',vmin=vmin,vmax=vmax)
    vmin,vmax = np.percentile(ysub,5.),np.percentile(ysub,90.)
    im3 = ax3.imshow(ysub,origin='lower',cmap='viridis',vmin=vmin,vmax=vmax)
    fig.colorbar(im3)

    fig = plt.figure()
    ax1 = fig.add_subplot(1,2,1)
    ax2 = fig.add_subplot(1,2,2)
    ax1.plot(ysub.sum(axis=0)),ax1.set_xlabel('pixX',fontsize=12),ax1.set_ylabel('cps',fontsize=12)
    ax1.plot(np.full_like(ysub.sum(axis=0),0.,dtype=float),':')
    ax2.plot(ysub.sum(axis=1)),ax2.set_xlabel('pixY',fontsize=12),ax2.set_ylabel('cps',fontsize=12)
    ax2.plot(np.full_like(ysub.sum(axis=1),0.,dtype=float),':')
    
```

```python
##########
# make different mask size
##########
halfdy_max = 10
halfdy_list = np.arange(halfdy_max)
objmask_dict = {}
for ii,i in enumerate(config):
    ny,nx = bkgobj[i].poly2d.data['Y'].shape
    trace = grismobj[i].output
    bbcorner = cutoutobj[i].bbcorner
    tx = trace['xh'] + trace['xyref'][0] - bbcorner['bb0x']
    ty = trace['yh'] + trace['xyref'][1] - bbcorner['bb0y']
    objmask_list = []
    for jj,j in enumerate(halfdy_list):
        t = ObjectMask(nx,ny,tx,ty,j,j)
        t.compute()
        objmask_list.append(t)
    objmask_dict[i] = objmask_list
```

```python
##### show #####
for ii,i in enumerate(config):
    for jj,j in enumerate(objmask_dict[i]):
        plt.figure()
        gdata = bkgobj[i].poly2d.data['Y']
        objmask_list = objmask_dict[i]
        m = np.isfinite(gdata)
        vmin,vmax = np.percentile(gdata[m],5.),np.percentile(gdata[m],90.)
        plt.imshow(gdata,origin='lower',cmap='viridis',vmin=vmin,vmax=vmax)
        plt.imshow(objmask_list[jj].mask,origin='lower',cmap='Greys',alpha=0.2)
        plt.title('{0} halfdy {1}'.format(i,jj))
    
```

```python
##########
# cps
##########
cps_dict = {}
for ii,i in enumerate(config):
    cps_list = []
    objmask_list = objmask_dict[i]
    for jj,j in enumerate(objmask_list):
        mdata = objmask_list[jj].mask.copy()
        gdata = bkgobj[i].poly2d.data['Y']
        bdata = bkgobj[i].poly2d.model['YFIT']
        exdata = (gdata - bdata) * mdata
        t = np.nansum(exdata,axis=0)[cutoutobj[i].bbox['padxleft']:-cutoutobj[i].bbox['padxright']]
        cps_list.append(t)
    cps_dict[i] = cps_list
##### show #####
for ii,i in enumerate(config):
    plt.figure(figsize=(10,10))
    objmask_list = objmask_dict[i]
    cps_list = cps_dict[i]
    for jj,j in enumerate(objmask_list):
        plt.plot(cps_list[jj],label='halfdy {0}'.format(halfdy_list[jj]))
        plt.legend(loc=(1.01,0.),fontsize=10)
        plt.xlabel('pixX',fontsize=12)
        plt.ylabel('cps',fontsize=12)
    plt.title(i,fontsize=12)
```

# Note:
We choose halfdy = 4 for both epochs.


```python
apcorr_dict = {}
sensitivity_dict = {}
wwperpix_dict = {}
for ii,i in enumerate(config):
    objmask_list = objmask_dict[i]
    trace = grismobj[i].output
    ##########
    # apcorr
    ##########
    instrument = 'HST-WFC3-IR-G141'
    apcorr_list = []
    for jj,j in enumerate(objmask_list):
        apsizepix = np.full_like(trace['ww'],halfdy_list[jj]*2 + 1,dtype=float)
        apsizearcsec = apsizepix * GrismApCorr().table[instrument]['scale']
        t = GrismApCorr(instrument=instrument,apsize=apsizearcsec,wave=trace['ww'],aptype='diameter',apunit='arcsec',waveunit='A')
        t.compute()
        apcorr_list.append(t.data['apcorr'])
    apcorr_dict[i] = apcorr_list
    ##########
    # sensitivity
    ##########
    sfile = ('/Users/kbhirombhakdi/_work/_calib_files/WFC3.IR.G141.cal.V4.32/WFC3.IR.G141.1st.sens.2.fits',1)
    sdata = pd.DataFrame(fits.open(sfile[0])[sfile[1]].data)
    smodel = interp1d(sdata.WAVELENGTH,sdata.SENSITIVITY,kind='linear',bounds_error=False,fill_value=None)
    sensitivity = smodel(trace['ww'])
    sensitivity[~np.isfinite(sensitivity)] = 1. # replace nan with 1. to avoid divided-by-zero problem
    sensitivity_dict[i] = sensitivity
    ##########
    # wwperpix
    ##########
    xhdiff = np.diff(trace['xh'])
    xhdiff = np.append(xhdiff,xhdiff[-1]) # add one last element to maintain the array dimension
    wwdiff = np.diff(trace['ww'])
    wwdiff = np.append(wwdiff,wwdiff[-1]) # add one last element to maintain the array dimension
    wwperpix = wwdiff/xhdiff
    wwperpix_dict[i] = wwperpix
```

```python
##########
# drzblot apcorr
##########
drzblot_apcorr_dict = {}
for ii,i in enumerate(config):
    objmask_list = objmask_dict[i]
    drzblot_apcorr_list = []
    for jj,j in enumerate(objmask_list):
        th = min(halfdy_list[jj],7)
        string = '../../GRB190829A/04_drzblot_crclean_apcorr/drzblot_S01_{0}_G141/drzblot_S01_{0}_G141_halfdy{1}_apcorr.csv'.format(i,th)
        t = pd.read_csv(string)
        model = interp1d(t.ww.values,t.apcorr.values,kind='nearest',bounds_error=False,fill_value='extrapolate')
        tt = model(trace['ww'])
        drzblot_apcorr_list.append(tt)
    drzblot_apcorr_dict[i] = drzblot_apcorr_list
    ##### show #####
    plt.figure()
    for jj,j in enumerate(objmask_list):
        plt.plot(trace['ww'],drzblot_apcorr_list[jj],label='halfdy {0}'.format(halfdy_list[jj]))
    plt.legend(loc=(1.01,0.))
    
```

```python
flam_dict = {}
for ii,i in enumerate(config):
    flam_list = []
    objmask_list = objmask_dict[i]
    for jj,j in enumerate(objmask_list):
        factor = apcorr_dict[i][jj] * drzblot_apcorr_dict[i][jj] * sensitivity_dict[i] * wwperpix_dict[i]
        flam = cps_dict[i][jj] / factor
        flam_list.append(flam)
    flam_dict[i] = flam_list
    
```

```python
wavebound = {'G141':(11000.,17000.)}
for ii,i in enumerate(config):
    objmask_list = objmask_dict[i]
    trace = grismobj[i].output
    flam_list = flam_dict[i]
    fig, ax1 = plt.subplots(figsize=(10,10))
    for jj,j in enumerate(objmask_list):
        x = trace['ww']
        y = flam_list[jj]
        m = np.argwhere((x >= wavebound['G141'][0])&(x <= wavebound['G141'][1])).flatten()
        ax1.plot(x, y,label='halfdy {0}'.format(halfdy_list[jj]))
    ax1.set_xlim(x.min(), x.max())
    ax1.set_ylim(1e-18,2e-17)
    ax1.set_xlabel('obs. wavelength (A)',fontsize=12)
    ax1.set_ylabel('flam',fontsize=12)
    ax2 = ax1.twiny()                    # make another frame that shares y-axis
    ax2.set_xlim(0,len(x)) # set the y-range limit
    ax2.set_title('pixX',fontsize=12)
    fig.legend(fontsize=10)
    fig.tight_layout()
```

```python
for ii,i in enumerate(config):
    objmask_list = objmask_dict[i]
    trace = grismobj[i].output
    for jj,j in enumerate(objmask_list):
        ww = trace['ww']
        cps = cps_dict[i][jj]
        apcorr = apcorr_dict[i][jj]
        drzblot_apcorr = drzblot_apcorr_dict[i][jj]
        sensitivity = sensitivity_dict[i]
        wwperpix = wwperpix_dict[i]
        flam = flam_dict[i][jj]
        output = {'ww':ww,'cps':cps,'apcorr':apcorr,'drzblot_apcorr':drzblot_apcorr,'sensitivity':sensitivity,'wwperpix':wwperpix,'flam':flam}
        string = './X01_{0}_G141/flam_halfdy{1}.csv'.format(i,halfdy_list[jj])
        pd.DataFrame(output).to_csv(string)
        print('Save {0}'.format(string))
    
```
