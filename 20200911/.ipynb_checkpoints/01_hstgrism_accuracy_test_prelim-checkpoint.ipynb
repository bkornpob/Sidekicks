# Story
HSTGRISM v8 is a python3 software for reducing grism images optimized for HST observations including WFC3/IR G102 and G141. We would like to validate its accuracy.

# Previously
[ISR WFC3 2016-15](https://www.stsci.edu/files/live/sites/www/files/home/hst/instrumentation/wfc3/documentation/instrument-science-reports-isrs/_documents/2016/WFC3-2016-15.pdf) presented the most recent review on WFC3/IR G102 and G141 trace and wavelength calibration. Orders +1, -1, and 2 were measured. The accuracy was better than 0.1 pixel for trace description, and 0.5 pixel for wavelength. Order 0th re-calibrated trace was now better than 1 pixel. VY2-2 was used as the calibrator. Only non-dithered observations of F098M/G102 and F140W/G141 were included from proposal IDs 11937, 12356, 12703, 13580, and 14023.

# Tasks
By following ISR WFC3 2016-15,
- Load 11937, 12356, 12703, 13580, and 14023
- Choose non-dithered F098/G102 and F140W/G141
- Extract 1D spectrum of each pair
- Measure the deviation of lines, and compute statistics
- Focus on 1st order.

# Resources
- VY2-2




```python
from astroquery.mast import Observations
import numpy as np
PID_list = ['14024','14023','14022',
            '13580','13579'
           ]
```

# Query and download from MAST

```python
Observations.get_metadata('observations')['Column Name'].tolist()
```

```python
for i in PID_list:
    obs_table = Observations.query_criteria(proposal_id=i,
                                            obs_collection='HST',
                                            filters=['F098M','G102'],
                                           )
    t = Observations.get_product_list(obs_table)
    Observations.download_products(t,
                                   download_dir='/Users/kbhirombhakdi/_work/_data/{0}/'.format(i),
                                   productSubGroupDescription='FLT'
                                  )
    
```

# Get hsum

```python
from astropy.io import fits
from hstgrism.headersummary import HeaderSummary
from hstgrism.container import Container
import pandas as pd
import glob,copy
```

```python
for j in PID_list:
    t = glob.glob('/Users/kbhirombhakdi/_work/_data/{0}/*/*/*/*flt.fits'.format(j))
    hsum_obj_list = []
    for ii,i in enumerate(t):
        tt = (i,1)
        hsum_obj = HeaderSummary(tt)
        hsum_obj.fetch()
        hsum_obj_list.append(hsum_obj)
    hsum_obj_dict = HeaderSummary(None).combine(hsum_obj_list)
    tt = pd.DataFrame(hsum_obj_dict).sort_values(by=['EXPSTART'])
    string = 'pid{0}'.format(j)
    cobj = Container(string,string,'pdf',False)
    string = './{0}/{0}_hsum.csv'.format(string)
    tt.to_csv(string)
    print('Save {0}'.format(string))
```

# Get xyd

```python
import matplotlib.pyplot as plt
from astropy.wcs import WCS
from astropy.coordinates import SkyCoord
from astropy.wcs.utils import skycoord_to_pixel
```

```python
xyd_dict = {'13580':[]}
for ii,i in enumerate(pair_dg_dict):
    hsum = pd.read_csv(hsum_file_dict[i])
    for jj,j in enumerate(pair_dg_dict[i]):
#         if jj!=0:
#             continue
        dfile = (hsum[hsum['index']==j[0]].filepath.values[0],1)
        gfile = (hsum[hsum['index']==j[1]].filepath.values[0],1)
        t = fits.open(dfile[0])
        tdata = t[dfile[1]].data
        RADEC = (t[0].header['RA_TARG'],t[0].header['DEC_TARG'])
        w = WCS(t[dfile[1]])
        c = SkyCoord(*RADEC,frame='icrs',unit='deg')
        xyd = skycoord_to_pixel(c,w,origin=0)
        fig = plt.figure(figsize=(3,3))
        ax1 = fig.add_subplot(111,projection=w)
        ax1.imshow(tdata,origin='lower',cmap='viridis',vmin=0,vmax=10)
        ax1.scatter(*xyd,s=30,edgecolor='red',facecolor='None')
        ax1.set_xlabel('RA')
        ax1.set_ylabel('DEC')
        string = hsum[hsum['index']==j[0]].ROOTNAME.values[0]
        ax1.set_title(string,fontsize=12)
        xyd_dict[i].append(xyd)
        
```

# HSTGRISM Flow

```python
from hstgrism.computexyref import ComputeXYREF
from hstgrism.wfc3irg102 import WFC3IRG102
from hstgrism.show_overview import show_overview
confile = '/Users/kbhirombhakdi/_work/_calib_files/WFC3.IR.G102.cal.V4.32/G102.F098M.V4.32.conf'
beam = 'A'
```

```python
container_dict = {'13580':[]}
for i in pair_dg_dict:
    hsum = pd.read_csv(hsum_file_dict[i])
    for j in pair_dg_dict[i]:
        sp = hsum[hsum['index']==j[1]].ROOTNAME.values[0]
        sf = sp
        t = Container(sp,sf,'pdf',False)
        container_dict[i].append(t)
        
```

```python
xyref_obj_dict = {'13580':[]}
for i in pair_dg_dict:
    for jj,j in enumerate(pair_dg_dict[i]):
        dfile = (hsum[hsum['index']==j[0]].filepath.values[0],1)
        gfile = (hsum[hsum['index']==j[1]].filepath.values[0],1)
        cobj = container_dict[i][jj]
        t = ComputeXYREF(xyd_dict[i][jj],'default','default',confile,beam,gfile,dfile,cobj)
        t.compute()
        t.save()
        xyref_obj_dict[i].append(t)
        
```

```python
grism_obj_dict = {'13580':[]}
for i in pair_dg_dict:
    for jj,j in enumerate(pair_dg_dict[i]):
        dfile = (hsum[hsum['index']==j[0]].filepath.values[0],1)
        gfile = (hsum[hsum['index']==j[1]].filepath.values[0],1)
        cobj = container_dict[i][jj]
        xyref = xyref_obj_dict[i][jj].data['xyref']
        t = WFC3IRG102(confile,beam,xyref,container=cobj)
        t.compute()
        t.save()
        grism_obj_dict[i].append(t)
        xyd = xyd_dict[i][jj]
        xh = t.output['xh']
        yh = t.output['yh']
        ww = t.output['ww']
        objname = 'VY2-2'
        save = True
        show_overview(gfile,dfile,xyd,xyref,xh,yh,ww,objname,save,container=cobj)
        
```
