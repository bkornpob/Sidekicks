# Story
We would like to measure accuracy of G102 wavelength map with HSTGRISM.

# Previously
20200911/01-02 selected dataset (i.e., Vy2-2) and extracted spectra (in cps unit). We need the standard lines of Vy2-2, and a routine to detect lines from the extracted spectra. Then, we can measure the accuracy.

# Resources
- [ISR WFC3-2009-18](https://www.stsci.edu/files/live/sites/www/files/home/hst/instrumentation/wfc3/documentation/instrument-science-reports-isrs/_documents/2009/WFC3-2009-18.pdf) for Vy2-2 line profile (see Table 3).
- Vy2-2 is local with v ~71.2 km/s ([Simbad](http://simbad.u-strasbg.fr/simbad/sim-id?Ident=PN%20Vy%202-2#lab_notes)).


```python
from specutils.fitting import find_lines_derivative
from specutils import Spectrum1D
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from hstgrism.container import Container
import astropy.units as u
VY22_LINE = {'[ArIII]':7751.1,'OI':8446.4,'HP11':8862.8,'[SIII]':9068.6,
             'HP9':9229.0,'[SIII]':9530.6,'HP7':10049.4,'[SII]':10321.9,'HeI':10830.3,
            }
```

```python
# Description:
#  - given VY22_LINE[k]
#  - find j that argmin(|lines_list[i][j][0].value - VY22_LINE[k]|)
#  - compute delta = |lines_list[i][j][0].value - VY22_LINE[k]|
#  - record k,j,delta
def compute_delta(VY22_LINE,lines):
    out = []
    for k in VY22_LINE:
        deltas = np.abs(lines - VY22_LINE[k])
        m = np.argwhere(deltas == deltas.min()).flatten()
        out.append((k,lines[m],deltas[m]))
    return out
```

```python
PID = '13580'
OBJ = 'VY2-2'
hsum = pd.read_csv('./pid{0}/pid{0}_hsum.csv'.format(PID))
pair_dg_list = [(5,17),(7,2),(4,14)]
confile = '/Users/kbhirombhakdi/_work/_calib_files/WFC3.IR.G102.cal.V4.32/G102.F098M.V4.32.conf'
beam = 'A'
container_list = []
for i in pair_dg_list:
    sp = hsum[hsum['index']==i[1]].ROOTNAME.values[0]
    sf = sp
    t = Container(sp,sf,'pdf',False)
    container_list.append(t)
```

```python
cps_list = []
for ii,i in enumerate(pair_dg_list):
    string = './{0}/{1}_cps.csv'.format(container_list[ii].data['savefolder'],container_list[ii].data['saveprefix'])
    t = pd.read_csv(string)
    cps_list.append(t)
```

```python
lines_list = []
for ii,i in enumerate(pair_dg_list):
    cps = cps_list[ii]
    x = cps.ww.values
    y = cps.cps.values
    spectrum = Spectrum1D(flux=y*u.Jy, spectral_axis=x*u.AA)
    flux_threshold = 0.75
    t = find_lines_derivative(spectrum, flux_threshold=flux_threshold)
    lines_list.append(t)
    
```

```python
line_match_list = []
for ii,i in enumerate(pair_dg_list):
    lines = lines_list[ii]['line_center'].value
    t = compute_delta(VY22_LINE,lines)
    line_match_list.append(t)
    
```

```python
lines_list[0]['line_center'].value
```

```python
for ii,i in enumerate(pair_dg_list):
    plt.figure()
    cps = cps_list[ii]
    lines = lines_list[ii]
    cps_min,cps_max = cps.cps.values.min(),cps.cps.values.max()
    tx,ty = cps.ww.values,cps.cps.values
    ty = (ty - cps_min)/cps_max
    plt.plot(tx,ty)
    lines = lines_list[ii]['line_center'].value
    for j in lines:
        tx = j
        plt.plot([tx,tx],[0,1],'kx')
    plt.plot([tx,tx],[0,1],'kx',label='Lines identified from spectrum')
    lines = line_match_list[ii]
    for j in lines:
        tx = j[1][0]
        string = 'delta = {0:.2f} A, {1}'.format(j[2][0],j[0])
        tl = plt.plot([tx,tx],[0,1],'x:',label=string)
        cc = tl[0].get_color()
        tx = VY22_LINE[j[0]]
        plt.plot([tx,tx],[0,1],color=cc,ls='-')
    plt.yscale('log')
    rootname = hsum[hsum['index']==i[1]].ROOTNAME.values[0]
    plt.title(rootname,fontsize=12)
    plt.xlabel('obs. wavelength (A)',fontsize=12)
    plt.ylabel('cps * scale',fontsize=12)
    plt.legend(loc=(1.01,0.),fontsize=10)
    string = './{0}/{1}_delta.pdf'.format(container_list[ii].data['savefolder'],container_list[ii].data['saveprefix'])
    plt.savefig(string,plotformat='pdf',bbox_inches='tight')
    print('Save {0}'.format(string))
    
```

```python
t = np.diff(cps.ww.values).mean()
print('G102 resolution is {0:.2f} A/pix'.format(t))
```

```python
keep = []
for ii,i in enumerate(line_match_list):
    t = line_match_list[ii]
    keep.append(pd.DataFrame(t)[2].values)
median = np.median(keep)
print('Median of absolute deviation = {0:.2f} A'.format(median[0]))
    
```

# Summary
We showed that HSTGRISM can map G102 with median of absolute deviation at about 13.73 A. We note that:
- ich501b1q having high deviation compared to the others was likely due to locating xyd.
- In any object that shows deviation larger than ~24 A (i.e., ~1 pixel in G102), it was likely due to wrong matching between the standard lines and the identified ones.
